### Step-1: Load data into a List

```python
import numpy as np
import time
data = np.loadtxt("input-programming-1.txt")
dataint = data.astype(np.int32)
num_list = np.array(dataint).tolist()
```

### Step-2: Bubble Sort algorithm and capturing Run Times (3 times per input, 20 inputs)

```python
def bubble_sort(A):
    for i in range(len(A)):
        for j in range(len(A)-1, i, -1):
            if A[j] < A[j-1]:
                A[j] , A[j-1] = A[j-1], A[j]
    return A

duration = []
inp_count = []
run_time, avg_time = 0, 0

for n in range(20):
    for t in range(3):
        inp_list = num_list[:500 * (n+1)]
        start = time.process_time()
        t_sort = bubble_sort(inp_list)
        run_time += (time.process_time() - start)
    avg_time = round(run_time / 3, 4)
    inp_count.append(len(inp_list))
    duration.append(avg_time)
    
print("Run Times:",  duration)
```

### Step-3: Plotting Input count vs Run Time

```python
import matplotlib.pyplot as plt
plt.plot(inp_count, duration, '-o')
plt.title('Input Count vs Run Time')
plt.xlabel('Input Counts')
plt.ylabel('Run Time (secs)')
plt.xticks(inp_count, rotation=90)
plt.show()
```
